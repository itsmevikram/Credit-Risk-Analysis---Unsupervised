# Replacing the Nulls as -1 for imputation purpose
# Applying imputation technique to replace these -1s with most frequent categories in the feature

```python
import numpy as np
raw_data.replace(np.nan,-1,inplace=True)
```

```python
from sklearn.impute import SimpleImputer
nan_categories=['Checking account','Saving accounts']
for nans in nan_categories:
    raw_data[nans] =  raw_data[nans].astype('category')
imputer=SimpleImputer(missing_values=-1, strategy='most_frequent')
raw_data[nan_categories] = imputer.fit_transform(raw_data[nan_categories])
```

```python
cleaned_data=raw_data.copy()
```

# Performing label Encoding to category columns

```python
categories=[]
for cat_cols in cleaned_data.columns:
    if cleaned_data[cat_cols].dtype=='object':
        categories.append(cat_cols)
```

```python
from sklearn.preprocessing import LabelEncoder
encoder=LabelEncoder()
for label in categories:
    cleaned_data[label]=encoder.fit_transform(cleaned_data[label])
```

```python
cleaned_data.head()
```

# Since this is a dataset which does not have a target feature,the system has to learn by its own through determining and adapting according to the structural characteristics in the input patterns,hence before applying clustering technique to build training models we normalize the data and apply it.

```python
#Scaling
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaled_data=scaler.fit_transform(cleaned_data)
scaled_data.shape
```

# finding the best number of cluster using sum of squares technique

```python
from sklearn.cluster import KMeans
sse=[]
for k in range(1, 21):
    kmeans=KMeans(n_clusters=k)
    kmeans.fit(scaled_data)
    sse.append(kmeans.inertia_)
plt.style.use('fivethirtyeight')
plt.figure(figsize=(12,6))
plt.plot(range(1, 21),sse,marker='o')
plt.xticks(range(1, 21))
plt.xlabel("Number of K-Clusters")
plt.ylabel("SSE")
plt.show()
```
